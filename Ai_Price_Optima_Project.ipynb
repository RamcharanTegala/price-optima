```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame loaded from 'dynamic_pricing.csv'
df = pd.read_csv('dynamic_pricing.csv')


# --- Step 1: Visualize Outliers ---
# This helps you see the outliers before formally detecting them.

# Define the numerical columns you want to inspect
num_cols = [
    'Number_of_Riders', 'Number_of_Drivers', 'Number_of_Past_Rides',
    'Average_Ratings', 'Expected_Ride_Duration', 'Historical_Cost_of_Ride'
]

plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    plt.boxplot(df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.savefig('outlier_boxplots.png') # Saves the plot as an image
print("Boxplots showing outliers have been generated and saved as 'outlier_boxplots.png'.")


# --- Step 2: Detect and Report Outliers (IQR Method) ---
# This function identifies and reports the number of outliers in each column.

def detect_outliers_iqr(data, column):
    """Detects outliers in a given column using the IQR method."""
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]

    return outliers, lower_bound, upper_bound

print("\n--- Outlier Detection Report ---")
for col in num_cols:
    outliers, lb, ub = detect_outliers_iqr(df, col)
    print(f"Column '{col}':")
    print(f"  - Found {len(outliers)} outliers.")
    print(f"  - IQR bounds: Lower < {lb:.2f}, Upper > {ub:.2f}")
```

```python
# --- Feature Engineering with dynamic_pricing.csv ---

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("/content/dynamic_pricing.csv")

# ------------------- Feature Engineering -------------------

# 1. Competitor Price Index (normalize by avg market price)
# Formula: Competitor_Price_Index = Historical_Cost_of_Ride / mean(Historical_Cost_of_Ride)
df['Competitor_Price_Index'] = df['Historical_Cost_of_Ride'] / df['Historical_Cost_of_Ride'].mean()

# 2. Cost per Minute
# Formula: Cost_per_Min = Historical_Cost_of_Ride / Expected_Ride_Duration
df['Cost_per_Min'] = df['Historical_Cost_of_Ride'] / df['Expected_Ride_Duration']

# 3. Driver to Rider Ratio
# Formula: Driver_to_Rider_Ratio = Number_of_Drivers / Number_of_Riders
df['Driver_to_Rider_Ratio'] = df['Number_of_Drivers'] / df['Number_of_Riders']

# 4. Rider to Driver Ratio
# Formula: Rider_Driver_Ratio = Number_of_Riders / Number_of_Drivers
df['Rider_Driver_Ratio'] = df['Number_of_Riders'] / df['Number_of_Drivers']

# 5. Inventory Health Index (proxy for coverage)
# Formula: Inventory_Health_Index = Number_of_Drivers / (Number_of_Riders + 1)
df['Inventory_Health_Index'] = df['Number_of_Drivers'] / (df['Number_of_Riders'] + 1)

# 6. Loyalty Score (normalized past rides)
# Formula: Loyalty_Score = Number_of_Past_Rides / max(Number_of_Past_Rides)
df['Loyalty_Score'] = df['Number_of_Past_Rides'] / df['Number_of_Past_Rides'].max()

# 7. Peak Indicator
# Convert Time_of_Booking to hour (023)
df['Booking_Hour'] = pd.to_datetime(df['Time_of_Booking'], errors='coerce').dt.hour
# Formula: Peak = 1 if (79) or (1720), else 0
df['Peak'] = df['Booking_Hour'].apply(lambda x: 1 if (7 <= x <= 9) or (17 <= x <= 20) else 0)

# 8. Supply Tightness
# Formula: Supply_Tightness = Number_of_Riders / Number_of_Drivers
df['Supply_Tightness'] = df['Number_of_Riders'] / df['Number_of_Drivers']

# 9. Vehicle Factor
# Formula: Vehicle_Factor = mapping(Vehicle_Type)
vehicle_map = {'sedan': 1.0, 'suv': 1.2, 'luxury': 1.5}
df['Vehicle_Factor'] = df['Vehicle_Type'].str.lower().map(vehicle_map)

# 10. Baseline Price
# Formula: baseline_price = Historical_Cost_of_Ride
df['baseline_price'] = df['Historical_Cost_of_Ride']

# 11. Competitor Price (synthetic, assume 5% higher than baseline)
# Formula: competitor_price = Historical_Cost_of_Ride  1.05
df['competitor_price'] = df['Historical_Cost_of_Ride'] * 1.05

# 12. Completion Probability (proxy)
# Formula: p_complete = min(1, Number_of_Drivers / Number_of_Riders)
df['p_complete'] = (df['Number_of_Drivers'] / df['Number_of_Riders']).clip(upper=1)

# 13. Final Price (dynamic pricing model)
# Formula: price = baseline_price  (1 + 0.1  Supply_Tightness)  Vehicle_Factor
df['price'] = df['baseline_price'] * (1 + 0.1 * df['Supply_Tightness']) * df['Vehicle_Factor']


# ------------------- Output -------------------

engineered_cols = [
    'Competitor_Price_Index', 'Cost_per_Min', 'Driver_to_Rider_Ratio',
    'Rider_Driver_Ratio', 'Inventory_Health_Index', 'Loyalty_Score',
    'Peak', 'Supply_Tightness', 'Vehicle_Factor', 'baseline_price',
    'competitor_price', 'p_complete', 'price'
]

print(" Engineered Feature Columns:")
print(engineered_cols)

print("\n Sample Engineered Feature Values:")
print(df[engineered_cols].head(10))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# --- 1. Setup: Load Data and Create Engineered Features ---
# Set plotting style for better aesthetics
sns.set_style("whitegrid")

# Load the dataset
df = pd.read_csv("/content/dynamic_pricing.csv")

# Create a new DataFrame for our feature-engineered data
df_fe = df.copy()

# Re-create the necessary engineered features from previous steps
df_fe['Rider_Driver_Ratio'] = df_fe['Number_of_Riders'] / (df_fe['Number_of_Drivers'] + 0.01)
df_fe['Driver_to_Rider_Ratio'] = df_fe['Number_of_Drivers'] / (df_fe['Number_of_Riders'] + 0.01)
df_fe['Supply_Tightness'] = df_fe['Rider_Driver_Ratio'] # As defined by mentor
df_fe['Cost_per_Min'] = df_fe['Historical_Cost_of_Ride'] / (df_fe['Expected_Ride_Duration'] + 0.01)
df_fe['Inventory_Health_Index'] = df_fe['Number_of_Drivers'] / (df_fe['Number_of_Riders'] + 1)
df_fe['Loyalty_Score'] = df_fe['Number_of_Past_Rides'] / (df_fe['Number_of_Past_Rides'].max() + 1)
vehicle_map = {'Economy': 1.0, 'Premium': 1.5}
df_fe['Vehicle_Factor'] = df_fe['Vehicle_Type'].map(vehicle_map)

# Create the proxy/example columns mentioned in the analysis plan
df_fe['p_complete'] = (df_fe['Number_of_Drivers'] / (df_fe['Number_of_Riders'] + 0.01)).clip(upper=1.0)
df_fe['baseline_price'] = df_fe['Historical_Cost_of_Ride']
df_fe['competitor_price'] = df_fe['Historical_Cost_of_Ride'] * 1.05
df_fe['price'] = df_fe['baseline_price'] * (1 + 0.1 * df_fe['Supply_Tightness']) * df_fe['Vehicle_Factor']

print("--- Setup Complete. Starting Analysis ---")


# --- 5.1 Seasonality: Conversion & Price by Time Buckets ---
print("\n--- 5.1 Seasonality Analysis ---")
seasonality = (
    df_fe.groupby("Time_of_Booking")
        .agg(mean_conv=("p_complete","mean"),
             mean_price=("price","mean"),
             total_intents=("Number_of_Riders","sum"))
        .sort_values("mean_conv", ascending=False)
)
print("Seasonality Summary:")
print(seasonality)

# Plotting the results
plt.figure(figsize=(10, 6))
seasonality["mean_conv"].plot(kind="bar", color=sns.color_palette('magma'))
plt.title("Mean Conversion Probability by Time of Day", fontsize=16)
plt.xlabel("Time of Booking")
plt.ylabel("Mean Conversion Probability")
plt.xticks(rotation=45)
plt.show()


# --- 5.2 Competitor Pressure: Price Gap & Index ---
print("\n--- 5.2 Competitor Pressure Analysis ---")
df_fe["Competitor_Gap_pct"] = (df_fe["price"] - df_fe["competitor_price"]) / df_fe["competitor_price"] * 100.0

comp_pressure = (
    df_fe.groupby("Vehicle_Type")
        .agg(
            avg_gap_pct=("Competitor_Gap_pct","mean"),
            mean_comp_price=("competitor_price","mean"),
            mean_our_price=("price","mean"),
        )
        .sort_values("avg_gap_pct")
)
print("Competitor Pressure Summary:")
print(comp_pressure)


# --- 5.3 Inventory Health Bands & Conversion ---
print("\n--- 5.3 Inventory Health Analysis ---")
bins = [0, 0.5, 0.8, 1.2, 10]
labels = ["Very Tight","Tight","Balanced","Loose"]
df_fe["Inventory_Band"] = pd.cut(df_fe["Driver_to_Rider_Ratio"], bins=bins, labels=labels, include_lowest=True)

inv_summary = df_fe.groupby("Inventory_Band").agg(
    mean_conv=("p_complete","mean"),
    mean_price=("price","mean"),
    count=("Inventory_Band","size")
)
print("Inventory Health Summary:")
print(inv_summary)

# Plotting the results
plt.figure(figsize=(10, 6))
inv_summary["mean_conv"].plot(kind="bar", color=sns.color_palette('viridis'))
plt.title("Mean Conversion Probability by Inventory Band", fontsize=16)
plt.xlabel("Inventory Health")
plt.ylabel("Mean Conversion Probability")
plt.xticks(rotation=45)
plt.show()


# --- 6. Customer Segmentation (KMeans) ---
print("\n--- 6. Customer Segmentation using KMeans ---")
# Step 1: Select features
seg_features = [
    "Loyalty_Score","Rider_Driver_Ratio","Driver_to_Rider_Ratio",
    "Supply_Tightness","Cost_per_Min","Inventory_Health_Index",
    "Expected_Ride_Duration"
]
X = df_fe[seg_features].copy().fillna(df_fe[seg_features].median())

# Step 2: Scale features
scaler = StandardScaler()
Xs = scaler.fit_transform(X)

# Step 3: Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df_fe["segment_k4"] = kmeans.fit_predict(Xs)
print(f"\nSuccessfully created {df_fe['segment_k4'].nunique()} customer segments.")

# Step 4: Segment profiling
seg_profile = (
    df_fe.groupby("segment_k4")
        .agg(
            mean_conv=("p_complete","mean"),
            mean_price=("price","mean"),
            mean_cost=("Historical_Cost_of_Ride","mean"),
            mean_loyalty=("Loyalty_Score","mean"),
            mean_ratio=("Rider_Driver_Ratio","mean"),
            count=("segment_k4","size")
        )
        .sort_values("mean_conv", ascending=False)
)
print("\nCustomer Segment Profiles:")
print(seg_profile)

print("\n--- Analysis Complete ---")
```

```python
import pandas as pd
import numpy as np

# 1. Load the Dataset
df = pd.read_csv("/content/dynamic_pricing.csv")

# 2. Define the Baseline Recommendation Function
def get_baseline_price_recommendation(ride_details):
    """
    Calculates a recommended price based on a fixed set of business rules
    derived from our exploratory data analysis.

    Args:
        ride_details (pd.Series or dict): A single row of data for a ride request.

    Returns:
        float: The recommended price.
    """
    # Rule 1: Start with a base price calculated from ride duration.
    # From our analysis, duration is the strongest predictor of cost.
    # Let's set a base rate of $2.5 per minute.
    base_price = ride_details['Expected_Ride_Duration'] * 2.5

    # Rule 2: Adjust for Vehicle Type.
    # Premium vehicles get a 50% markup over the base.
    if ride_details['Vehicle_Type'] == 'Premium':
        price = base_price * 1.5
    else:
        price = base_price

    # Rule 3: Adjust for Time of Day (Seasonality).
    # Our analysis showed 'Evening' and 'Night' are peak times.
    if ride_details['Time_of_Booking'] in ['Evening', 'Night']:
        price *= 1.20  # Apply a 20% surcharge

    # Rule 4: Apply dynamic "surge" pricing based on Inventory Health.
    driver_to_rider_ratio = ride_details['Number_of_Drivers'] / (ride_details['Number_of_Riders'] + 0.01)

    if driver_to_rider_ratio < 0.5:  # 'Very Tight' supply
        price *= 1.40  # 40% surge
    elif driver_to_rider_ratio < 0.8:  # 'Tight' supply
        price *= 1.15  # 15% surge
    elif driver_to_rider_ratio > 1.2:  # 'Loose' supply (oversupply)
        price *= 0.90  # 10% discount

    # Finally, ensure a minimum price for any ride.
    final_price = max(price, 10.0)

    return round(final_price, 2)

# 3. Demonstrate the Function on Sample Rides
print("--- Demonstrating Baseline Recommender on Sample Scenarios ---")

# Scenario 1: High-demand, expensive ride
sample_high_demand = {
    'Expected_Ride_Duration': 45,
    'Vehicle_Type': 'Premium',
    'Time_of_Booking': 'Evening',
    'Number_of_Drivers': 15,
    'Number_of_Riders': 40  # 'Very Tight' supply
}
price_1 = get_baseline_price_recommendation(sample_high_demand)
print(f"Recommended Price for high-demand scenario: ${price_1}")

# Scenario 2: Low-demand, cheaper ride
sample_low_demand = {
    'Expected_Ride_Duration': 15,
    'Vehicle_Type': 'Economy',
    'Time_of_Booking': 'Afternoon',
    'Number_of_Drivers': 50,
    'Number_of_Riders': 30  # 'Loose' supply
}
price_2 = get_baseline_price_recommendation(sample_low_demand)
print(f"Recommended Price for low-demand scenario: ${price_2}")


# 4. Apply the Baseline Recommender to the Entire DataFrame
print("\n--- Applying recommendations to the full dataset... ---")
df['Baseline_Recommended_Price'] = df.apply(get_baseline_price_recommendation, axis=1)

# Display the results for comparison
print("\nComparison of Historical Cost vs. Baseline Recommended Price:")
print(df[['Historical_Cost_of_Ride', 'Baseline_Recommended_Price']].head())
```

```python
import pandas as pd
import numpy as np

# --- 1. Load Data and Define the Baseline Recommender ---

# Load the dataset
df = pd.read_csv("/content/dynamic_pricing.csv")

# Include the baseline recommender function from our previous step
def get_baseline_price_recommendation(ride_details):
    base_price = ride_details['Expected_Ride_Duration'] * 2.5
    price = base_price * 1.5 if ride_details['Vehicle_Type'] == 'Premium' else base_price
    if ride_details['Time_of_Booking'] in ['Evening', 'Night']:
        price *= 1.20
    driver_to_rider_ratio = ride_details['Number_of_Drivers'] / (ride_details['Number_of_Riders'] + 0.01)
    if driver_to_rider_ratio < 0.5: price *= 1.40
    elif driver_to_rider_ratio < 0.8: price *= 1.15
    elif driver_to_rider_ratio > 1.2: price *= 0.90
    return round(max(price, 10.0), 2)

# --- 2. Prepare the Two Scenarios for Comparison ---

# Scenario A: Baseline DataFrame
df_baseline = df.copy()
df_baseline['Price'] = df_baseline['Historical_Cost_of_Ride']
df_baseline['Cost'] = df_baseline['Historical_Cost_of_Ride']
df_baseline['Booking_Intents'] = 1
df_baseline['Completed_Rides'] = 1 # Assumption: all rides were completed at the historical price

# Scenario B: Recommended DataFrame
df_recommended = df.copy()
df_recommended['Price'] = df_recommended.apply(get_baseline_price_recommendation, axis=1)
df_recommended['Cost'] = df_recommended['Historical_Cost_of_Ride']
df_recommended['Booking_Intents'] = 1

# Simple simulation for completed rides based on price change
price_ratio = df_recommended['Price'] / df_baseline['Price']
p_complete = np.clip(np.exp(-0.5 * (price_ratio - 1)), 0.1, 1.0) # Price elasticity
df_recommended['Completed_Rides'] = (np.random.rand(len(df)) < p_complete).astype(int)


# --- 3. Define the KPI Calculation Function ---

def generate_kpi_summary(df_base, df_reco):
    """Calculates and formats a KPI summary table comparing two scenarios."""

    # --- Calculate KPIs for Baseline ---
    kpis_base = {}
    kpis_base['Revenue ($)'] = (df_base['Price'] * df_base['Completed_Rides']).sum()
    kpis_base['Rides (#)'] = df_base['Completed_Rides'].sum()
    kpis_base['Avg. Price ($)'] = kpis_base['Revenue ($)'] / kpis_base['Rides (#)']
    kpis_base['Profit ($)'] = ((df_base['Price'] - df_base['Cost']) * df_base['Completed_Rides']).sum()
    kpis_base['Gross Margin (%)'] = (kpis_base['Profit ($)'] / kpis_base['Revenue ($)']) * 100
    kpis_base['Cancellations (#)'] = 0 # By definition in baseline
    kpis_base['Cancellation Rate (%)'] = 0.0
    kpis_base['Price Changes (#)'] = 0 # No changes in baseline
    kpis_base['Price Change Rate (%)'] = 0.0

    # --- Calculate KPIs for Recommended ---
    kpis_reco = {}
    kpis_reco['Revenue ($)'] = (df_reco['Price'] * df_reco['Completed_Rides']).sum()
    kpis_reco['Rides (#)'] = df_reco['Completed_Rides'].sum()
    kpis_reco['Avg. Price ($)'] = kpis_reco['Revenue ($)'] / kpis_reco['Rides (#)'] if kpis_reco['Rides (#)'] > 0 else 0
    kpis_reco['Profit ($)'] = ((df_reco['Price'] - df_reco['Cost']) * df_reco['Completed_Rides']).sum()
    kpis_reco['Gross Margin (%)'] = (kpis_reco['Profit ($)'] / kpis_reco['Revenue ($)']) * 100 if kpis_reco['Revenue ($)'] > 0 else 0
    kpis_reco['Cancellations (#)'] = df_reco['Booking_Intents'].sum() - kpis_reco['Rides (#)']
    kpis_reco['Cancellation Rate (%)'] = (kpis_reco['Cancellations (#)'] / df_reco['Booking_Intents'].sum()) * 100
    kpis_reco['Price Changes (#)'] = (df_reco['Price'] != df_base['Price']).sum()
    kpis_reco['Price Change Rate (%)'] = (kpis_reco['Price Changes (#)'] / len(df_reco)) * 100

    # --- Create Final Report DataFrame ---
    report = pd.DataFrame({'baseline': kpis_base, 'recommended': kpis_reco})
    report['delta'] = report['recommended'] - report['baseline']
    report['delta_pct'] = (report['delta'] / report['baseline'].replace(0, 1)) * 100
    report.loc['Revenue Lift (%)'] = [None, None, None, report.loc['Revenue ($)', 'delta_pct']]

    # Formatting
    for col in ['baseline', 'recommended', 'delta']:
        for kpi in ['Revenue ($)', 'Profit ($)', 'Avg. Price ($)']:
            report.loc[kpi, col] = f"{report.loc[kpi, col]:,.2f}"
        for kpi in ['Rides (#)', 'Cancellations (#)', 'Price Changes (#)']:
            report.loc[kpi, col] = f"{int(report.loc[kpi, col]):,}"
    for col in ['baseline', 'recommended', 'delta_pct', 'delta']:
        for kpi in ['Gross Margin (%)', 'Cancellation Rate (%)', 'Price Change Rate (%)', 'Revenue Lift (%)']:
             if pd.notna(report.loc[kpi, col]):
                report.loc[kpi, col] = f"{report.loc[kpi, col]:.2f}"

    return report.fillna('')

# --- 4. Generate and Display the Report ---
kpi_report = generate_kpi_summary(df_baseline, df_recommended)

print("--- Baseline vs Recommended KPIs ---")
print(kpi_report)
```
